```python
import random

def get_name():
    names = ['Adam', 'Alan', 'Alex', 'Alvin', 'Andrew', 'Anthony', 'Arnold', 'Jin', 'Billy', 'Anchal']
    return random.choice(names)

def get_age(start=20, end=40):
    return np.random.randint(start, end + 1)
```

```python
ls = []
for r in range(8):
    ls.append({'UserID': r, 'Name': get_name(), 'Age': get_age()})

user_df = pd.DataFrame(ls)
user_df
```

```python
nl = []
while True:
    n = get_name()
    if  n not in nl:
        nl.append(n)
    if len(nl) == 8:
        break;
        
al = []
il = []
for r in range(8):
    al.append(get_age())
    il.append(r)

dic = {'UserID': il,  'Name': nl, 'Age':al}
user_df = pd.DataFrame(dic, columns = ['UserID', 'Name', 'Age'])
user_df

# 다른 방법으로 해보기
```

```python
user_df.UserID += 1
```

```python
user_df
```

```python
money_df = pd.DataFrame(columns  = ['ID', 'Money'])

for idx in range(15):
    money = np.random.randint(1,21) * 1000
    data = {'Money': money, 'ID': random.randint(1, 8)}
    money_df.loc[len(money_df)] = data

money_df

# df2['ID'].unique()
```

```python
result_df = pd.merge(user_df, money_df, how="right", left_on='UserID', right_on='ID')
result_df
# 다시
```

```python
money_list = result_df.groupby('Name')['Money'].agg(['sum']).reset_index().rename(columns={'sum':'Money'})
# money_list = result_df.groupby('Name').sum()['Money'].reset_index().rename(columns={'sum':'Money'})
money_list
```

```python
result = user_df.merge(money_list, how='left').fillna(0)
result
```

```python
# NaN to 0
# money float -> int
# data.pkl 로 저장

result['Money'] = pd.to_numeric(result['Money'], downcast='signed')
result
```
