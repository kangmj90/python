```python
import requests
from bs4 import BeautifulSoup

import pandas as pd
import numpy as np
```

```python
# guest
response = requests.get('https://clpapi.tinkeredu.co.kr/static/docs/guest.html')
response
```

```python
dom = BeautifulSoup(response.content, 'html.parser')
dom
```

```python
data = dom.select('table.tableblock tbody tr td:nth-child(3)')
len(data), data
```

```python
with open("guest.txt", "w") as outfile:
    outfile.write("\n".join(str(item) for item in data))
```

```python
# guets online degree
response = requests.get('https://clpapi.tinkeredu.co.kr/static/docs/teacher_mission.html')
dom = BeautifulSoup(response.content, 'html.parser')
data = dom.select('table.tableblock tbody tr td:nth-child(3)')

with open("teacher_mission.txt", "w") as outfile:
    outfile.write("\n".join(str(item) for item in data))
```
