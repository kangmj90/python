```python
import seaborn as sns

# https://github.com/mwaskom/seaborn-data
```

```python
iris = sns.load_dataset('iris')

iris.tail()
```

```python
sns.lmplot('sepal_length', 'sepal_width', data=iris, hue='species', fit_reg=False)
# linear model (선형회귀) + plot
# fit_reg : 선형회귀 T/F
plt.show()
```

```python
sns.lmplot('petal_length', 'petal_width', data=iris, hue='species', fit_reg=False)
plt.show()
```

```python
sns.pairplot(iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']], hue='species')
plt.show()
```
