```python
import pickle
import time
for i in range(len(lis)):
    print(i, end = '  ')
    driver.find_element_by_css_selector('div.depth > ul > li > a')

    response = driver.find_element_by_css_selector('.home_section').get_attribute('innerHTML')
    dom = BeautifulSoup(response, 'html.parser')

    dates_keywords = dom.select('.rank_inner')
    
    time.sleep(5)
    naver_df2 = pd.DataFrame(columns = ['date', 'rank', 'keyword'])
    for dates in dates_keywords[:-3]:
        date = dates.select_one('.title_cell').text
        keywords = dates.select('.list')

        for keyword in keywords:
            naver_df2.loc[len(naver_df2)] = {
                'date': date,
                'rank': int(keyword.select_one('.num').text),
                'keyword': keyword.select_one('.title').text
            }
            
    pivot_data = naver_df2.pivot('rank', 'date', 'keyword')
            
    with open('datas/datalab_' + str(i) + '.pkl', 'wb') as f:
        pickle.dump(pivot_data, f)
```

```python
def download(title, link):
#     stream > 연속적으로 !
    response = requests.get(link, stream=True)
    
    save_path, size = './datas/{}'.format(title), 0
    
    with open(save_path, 'wb') as f:
        for chunk in response.iter_content(chunk_size=1024):
            if chunk:
                size += 1024
                f.write(chunk)
                
    return size
```
