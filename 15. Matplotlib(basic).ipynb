## 4. 스타일
- https://matplotlib.org/examples/color/named_colors.html

색상
- blue : b
- green : g
- red : r
- cyan : c
- magenta : m
- yellow : y
- black : k
- white : w

마커
- . : point marker
- , : pixel marker
- o : circle marker
- v : triangle_down marker
- ^ : triangle_up marker
- < : triangle_left marker
- `>` : triangle_right marker
- 1 : tri_down marker
- 2 : tri_up marker
- 3 : tri_left marker
- 4 : tri_right marker
- s : square marker
- p : pentagon marker
- `*` : star marker
- h : hexagon1 marker
- H : hexagon2 marker
- `+` : plus marker
- x : x marker
- D : diamond marker
- d : thin_diamond marker

라인 스타일
- `-` : solid line style
- `--` : dashed line style
- `-.` : dash-dot line style
- `:` : dotted line style

커스텀 스타일
- c : 선 색상
- lw : 선 굵기
- ls : 선 스타일
- marker : 마커 종류
- ms : 마커 크기
- mec : 마커 선 색깔
- mew : 마커 선 굵기
- mfc : 마커 내부 색깔

단축 옵션으로 한번에 설정
- 색상, 마커, 라인 스타일을 한번에 설정가능

```python
data = np.random.randint(10, size=10)
plt.plot(data, 'bo--')
plt.show()
```

커스텀하게 설정
- 단축 옵션으로 설정하는것보다 더 많은 설정가능

```python
data = np.random.randint(10, size=10)
plt.plot(data, c="b", ls="--", marker="o", lw=5,
         ms="10", mec="r", mew="3", mfc="g")
plt.show()
```

## 5. limit 설정
- x축과 y축의 범위를 정의

xlim, ylim 설정

```python
# x 축 : -5 ~ 20 까지의 범위, y 축 : -10 ~ 30 까지의 범위
data = np.random.randint(10, size=10)
plt.xlim(-5, 20)
plt.ylim(-10, 30)
print(data)
plt.plot(data)
plt.show()
```

## 6. tick 설정
- 그래프에서의 x축과 y축의 값을 설정

```python
data = np.random.randint(5, size=5)
plt.plot(data)
plt.show()
```

## 8. 멀티 라인 그리기

하나의 plot 함수로 그리기

```python
data1 = np.random.randint(5, size=5)
data2 = np.random.randint(5, size=5)
data3 = np.random.randint(5, size=5)
xticks = range(0, 5)
print(data1, data2, data3)

plt.xticks(range(0, 5))
plt.plot(xticks, data1, 'ro-', xticks, data2, 'g^:', xticks, data3, 'b*--')
plt.show()
```

여러개의 plot 함수로 그리기

```python
xticks = range(0, 5)
print(data1, data2, data3)

plt.plot(xticks, data1)
plt.plot(xticks, data2)
plt.plot(xticks, data3)
plt.show()
```

## 9. legend (범례)
범례 위치
- best : 0
- upper right : 1
- upper left : 2
- lower left : 3
- lower right : 4
- right : 5
- center left : 6
- center right : 7
- lower center : 8
- upper center : 9
- center : 10

가장 적절한 위치에 자동으로 범례 배치 

```python
data1 = np.random.randint(5, size=5)
data2 = np.random.randint(5, size=5)
plt.plot(data1, label="data1")
plt.plot(data2, label="data2")
plt.legend(loc=0)
plt.show()
```

우측 상단에 범례 배치

```python
data1 = np.random.randint(5, size=5)
data2 = np.random.randint(5, size=5)
plt.plot(data1, label="data1")
plt.plot(data2, label="data2")
plt.legend(loc=1)
plt.show()
```

## 10. title 추가
- 그래프의 타이틀을 설정가능
- 기본 폰트가 한글폰트가 없는 `sans-serif`로 되어 있기 때문에 한글을 사용하면 글자가 깨짐.

```python
data = np.random.randint(5, size=5)
plt.plot(data)
plt.xlabel("xtick")
plt.ylabel("ytick")
plt.title("Title")
plt.show()
```

## 11. Figure
- 그래프의 크기를 조절
- 여러개의 그래프 출력

### 11.1 그래프 크기 조절

```python
data = np.random.randint(5, size=5)
plt.figure(figsize=(10, 3))  # (가로, 세로), 인치단위
plt.plot(data)
plt.show()
```

### 11.2 여러개 그래프 출력

```python
data1 = np.random.randint(5, size=5)
data2 = np.random.randint(5, size=5)
print(data1, data2)

plt.subplot(2, 1, 1)  # 2*1에서 1번 위치
plt.plot(data1)

plt.subplot(2, 1, 2)  # 2*1에서 2번 위치
plt.plot(data2)

plt.show()
```

```python
# subplot 인자 다르게주기
data = [np.random.randint(5, size=5) for _ in range(4)]

print(data)

for idx in range(4):
    print("2*2에서 {}번 위치 : {}".format(idx + 1, 221 + idx))
    plt.subplot(221 + idx)
    plt.plot(data[idx])

plt.show()
```
