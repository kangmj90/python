### 6.1. 붓꽃(iris) 데이터
- 대망의 아이리스 데이터 셋
- 데사한다고 하고 아이리스 데이터를 한번도 못봤다고하면 정말 말도 안되는 것
- 세 종류의 아이리스 종(setosa, virginica, versicolor)의 sepal(꽃받침), petal(꽃잎)의 length(길이), width(폭)을 정리한 데이터(5컬럼)
- 변수: sepal_length, sepal_width, petal_length, petal_width
- label: species
- classification의 대표적인 데이터로, 아주 기본적인 ML에서 예제로 사용
- 여기서는 시각화의 예제로 활용

```python
iris = sns.load_dataset("iris")
iris.tail()
```

#### 꽃받침의 길이와 넓이로 종을 구분해보자!

```python
sns.lmplot("sepal_length", "sepal_width", data=iris, hue='species', fit_reg=False)
# lmplot = lm(linear model; 선형회귀)+plot
# 원래 목적은 R 스타일 선형회귀 그래프를 그리는 것이나, 여기서는 이쁘니까 그냥 fit_reg를 False로 주고 쓴다.
# fit_reg=True면 각 데이터에 대해 선형회귀 그래프를 그려버린다...
plt.show()
```

- setosa 구분은 가능하지만 versicolor, virginica는 구분이 어렵당..

#### 그럼 이번엔 꽃잎의 길이와 넓이로 종을 구분해보자!

```python
sns.lmplot("petal_length", "petal_width", data=iris, hue='species', fit_reg=False)
plt.show()
```

- setosa 구분이 확실하게 가능하고 versicolor, virginica의 구분도 어느정도 가능함을 알 수 있다.
- 꽃받침으로 종을 구분하는것보다 꽃잎으로 종을 구분하는것이 더 쉬울 것이라는 결론에 도달!
- 이것이 바로 시각화
    - 여기서 발전된 개념이 EDA(Explorative Data Analysis)

#### EDA(Explorative Data Analysis; 탐색적 데이터 분석)
- 쉽게 말해 시각화로 그냥 이렇게 저렇게 그래프 그려봐서 데이터가 어떻게 구성되어있나 살펴보는 것!
- 처음엔 고전 통계학에서 굉장히 '하등한 것'으로 취급되었지만, 요새 데-사- 에서는 아주 중요한 영역이다.
    - 데이터가 어떻게 생겨먹었는지를 봐야 어떻게 분석할지 감이 오지 않겠는가!
    - 바로 이번의 예제에서 처럼!

#### 꽃잎으로 구분하는 것이 쉽다는 것을 알았으니 이제 pair plot을 그려보자
- pair plot: x위치와 y위치에 각자 변수를 놓고 각 변수끼리의 2차원 그래프를 여러개 그리는 것

```python
sns.pairplot(iris[["petal_length", "petal_width", "species"]], hue="species")
plt.show()
```

### 6.2 팁 데이터
- 변수: total_bill(총 금액), sex(성별), smoker(흡연자), day(요일), time(식사시간대), size(식사인원)
- label: tip
- 변수로 라벨을 예측하기(회귀)

```python
tips = sns.load_dataset("tips")
tips.tail(5)
```

#### 6.2.1 식사시간과 성별에 따른 팁의 총량을 시각화
- parameter
    - annot : 수치 정보 출력
    - fmt : 수치데이터 포멧 ".2f" 소수 둘째 자리
    - as_cmap: 스무스하게 표현
- [light_palette](http://seaborn.pydata.org/generated/seaborn.light_palette.html)

- 시간별 성별 팁의 총량

```python
tips_sum = tips.pivot_table(index="time", columns="sex", values="tip", aggfunc=np.sum)
tips_sum
```

##### * insight: 점심시간 보다 저녁시간에 팁이 많고, 저녁시간에 남자손님에게 받는 팁이 가장 많다.

```python
# 위의 DataFrame을 보기좋게 heatmap스타일로 시각화
sns.heatmap(tips_sum, cmap=sns.light_palette("purple", as_cmap=True), annot=True, fmt=".2f")
plt.show()
```

#### 6.2.2 날짜와 흡연여부에 따른 팁을 주는 정도
- 날짜와 흡연여부에 따라서 가장 효율이 좋은 테이블을 선점하여 서빙을 하는게 좋을까?

- 날짜와 흡연여부에 따른 팁의 총 액수

```python
tips_amount = tips.pivot_table(index="smoker", columns="day", values="tip", aggfunc="sum")
tips_amount
```

- 날짜와 흡연여부에 따른 고객 수

```python
tips_count = tips.pivot_table(index="smoker", columns="day", values="tip", aggfunc="count")
tips_count
```

- 날짜와 흡연여부에 따른 평균 팁

```python
tips_result = round(tips_amount / tips_count, 2)
tips_result
```

##### * insight: 일요일에 담배를 피우는 사람이 팁을 가장 많이 준다.

#### 그렇다면 일요일 저녁? 점심?

```python
tips[tips["day"]=="Sun"].groupby("time").sum()
```

##### * insight: 일요일에는 점심을 제공하지 않는다.

### 결론: 일요일 저녁에 담배를 피우는 사람의 테이블을 맡으면 적게 일하고 많은 팁을 받을수 있다.

```python
sns.heatmap(tips_result, cmap=sns.light_palette("seagreen", as_cmap=True), annot=True, fmt=".2f")
plt.show()
```

#### 6.2.3 전체 계산 금액과 팁의 상관관계

```python
import scipy.stats as stats
s = sns.jointplot("total_bill", "tip", data=tips)
pearsonr, p_value = stats.pearsonr(tips["total_bill"], tips["tip"])
print("pearsonr : {}, p_value : {}".format(pearsonr, p_value))
plt.show()
```

- pearson 상관계수 0.68이며, p-value가 0.05보다 작기 때문에, 우리는 전체 계산 금액과 팁이 상관성이 있다는 사실을 알 수 있다.
- 단 여기서 '상관관계(연관성;association)'이지, '인과관계(인과성;causation)'가 아니다. (팁을 많이 내기 때문에 전체적으로 팁이 아까워 많이 먹는지, 많이 먹었으니까 팁을 많이 내는지, 돈을 많이 내는 부자라 팁도 많이 주는지 알 수 없다.)
- 상관관계는 인과관계의 필요조건

##### 상관계수(correlation coefficient)
- 얼마나 연관이 되어 있는가를 숫자로 알 수 있는 방법
- 두 변수 간의 강도를 숫자로 표현하는 방법
- 두 변수 사이의 직선적인 경향이 얼마나 강할까?
- 완전/직선은 상관계수 1 (강한 양의 상관관계)
- 무상관 0
- 완전\직선은 상관계수 -1 (강한 음의 상관관계)
- in graph?

##### 1종 오류, 2종 오류
||진짜 참|진짜 거짓|
|---|---|---|
|<b>주장 참</b>|진실을 밝혔다|개사기꾼(1종오류(α))|
|<b>주장 거짓</b>|장님(2종오류(β))|진실을 밝혔다

##### p-value(유의확률), α(유의수준), H<sub>0</sub>(귀무가설), H<sub>a</sub>(대립가설)
- H<sub>0</sub>(귀무가설)
    - 내가 주장하고 싶은 '가설'이 아무 의미가 없을 때의 가설
    - 가령 나는 'A와 B가 다르다'고 주장하고 싶으면, 여기서 귀무가설은 내 주장이 아무 의미가 없을 때의 가설인 'A와 B가 같다'
- H<sub>a</sub>(대립가설)
    - 내가 주장하고 싶은 가설
- p-value(유의확률)
    - 귀무가설이 참이라고 가정했을 때 그 변수가 가질 수 있는 값의 확률 분포에서 어떤 검정통계량이 발생할 확률
- a(유의수준)
    - 1종 오류의 최대 허용 한계
    - 우연히 일어날 확률
    
- 즉, 귀무가설이 참이라고 가정했을 때 그 변수가 가질 수 있는 값의 확률 분포에서 어떤 검정통계량이 발생할 확률이 우연히 일어날 확률 보다 낮을 때 귀무가설을 기각할 수 있다. (if p-value < 0.05 then we can reject null hypothesis)

##### Example

1. H0: u = 80, Ha: u =/= 80 [양측검정]
2. (정규성 가정->안되면 t등) u=80이 평균인 정규분포곡선을 그린다.
3. 당연히 귀무가설이 참이라면 관측된 통계값 u는 80근처에서 가장 확률이 높을 것이다.
4. 그러나 그 통계"값"이 우연히 일어날 확률(유의확률)에 해당하는 "값"보다 커지면(통계값의 발생 확률이 우연히 발생할 확률보다 작아지면)
5. 이 검정통계량은 우연에 의해서 생긴 결과가 아니라고 말할 수 있으며
6. 애초의 초기 가정인 귀무가설을 기각할 수 있다.

[사실 우연히 일어날 수 있는 일이었는데, 우연히 일어난 것이 아니라 뭔가 효과가 있어서 일어난 일이라고 주장 => 1종 오류 발생 => 여기서 '우연히'에 대해 얼만큼의 확률을 적용할 것인지 => '1종 오류의 최대 허용 한계'설정 => 유의수준 = 1종 오류가 발생할 확률의 최댓값]
